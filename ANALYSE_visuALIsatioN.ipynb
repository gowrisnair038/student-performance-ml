```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['exam_score'])
plt.title("Distribution of exam scores")
plt.xlabel("Exam Score")
plt.ylabel("No:of students")
plt.show()
```

```python
plt.scatter(df['study_hours'],df['exam_score'],s=10,alpha=0.3)
plt.title("Study hours vs Exam scores")
plt.xlabel("Study hours")
plt.ylabel("Exam score")
plt.show()
```

```python
sns.boxplot(x='gender',y='exam_score',data=df)
plt.title("Exam score comparison by gender")
plt.show()
```

```python
sns.barplot(x='study_method',y='exam_score',data=df)
plt.title("Study methods vs exam scores")
plt.show()
```

```python
#Conclusion
#Exam performance increases with study hours, but the relationship is not perfectly strong.
#Exam scores are nearly equally distributed across genders, indicating no significant gender-based performance gap.
#Study method affects performance, and structured learning methods like coaching seem to produce better results.
```

```python
#From the visualizations, we can conclude that:
#	•	Study hours have a positive effect on exam scores.
#	•	Gender does not significantly impact exam performance.
#	•	Study method plays an important role, with coaching and mixed methods leading to higher average scores.
```
